**DATA VISUALISATION**

```python
corr=data1.corr()
di,ax=plt.subplots(figsize=(7,5))
sns.heatmap(corr,annot=True)
```

```python
corr=data2.corr()
di,ax=plt.subplots(figsize=(7,5))
sns.heatmap(corr,annot=True)
```

```python
data1.columns= ["State","Date","Frequency","EUR","EE","ELPR","Region"]
plt.figure(figsize=(10,7))
plt.title("Unemployment Rate")
sns.histplot(x="EUR",hue="Region",data=data1)
```

```python
data2.columns= ["State","Date","Frequency","EUR","EE","ELPR","Region","longitude","latitude"]
plt.figure(figsize=(10, 8))
plt.title("Unemployment Rate")
sns.histplot(x="EUR", hue="Region", data=data2)
plt.show()
```

```python
ue= data2[["State", "Region", "EUR"]]

fig = px.sunburst(ue, path=['Region','State'], values='EUR',
                  title= 'Unemployment rate in every State and Region', height=650)
fig.show()
```
